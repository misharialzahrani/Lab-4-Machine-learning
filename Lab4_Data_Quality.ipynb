```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

df = pd.read_csv("Chocolate_Sales.csv")
df.head()
```

```python
print("Shape:", df.shape)

print("\nData Types BEFORE:\n")
print(df.dtypes)

print("\nMissing Values BEFORE:\n")
print(df.isna().sum())

print("\nDuplicate Rows BEFORE:", df.duplicated().sum())
```

```python
# Convert Date column
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')

# Clean Amount column (remove $ and commas)
df['Amount'] = df['Amount'].replace(r'[\$,]', '', regex=True).astype(float)

print("\nData Types AFTER:\n")
print(df.dtypes)
```

```python
# Check if date conversion created missing values
print("Missing values after conversion:\n")
print(df.isna().sum())

# Remove rows with missing Date if any
df = df.dropna(subset=['Date'])

print("\nMissing Values AFTER Cleaning:\n")
print(df.isna().sum())
```

```python
print("Duplicate Rows BEFORE:", df.duplicated().sum())

df = df.drop_duplicates()

print("Duplicate Rows AFTER:", df.duplicated().sum())
```

```python
plt.figure()
sns.boxplot(x=df['Amount'])
plt.title("Outlier Detection - Revenue")
plt.show()

plt.figure()
sns.boxplot(x=df['Boxes Shipped'])
plt.title("Outlier Detection - Boxes Shipped")
plt.show()
```

```python
print("Final Dataset Shape:", df.shape)

print("\nFinal Missing Values:\n")
print(df.isna().sum())

df.head()
```

In this lab, data quality assessment was performed on the Chocolate Sales dataset. Missing values and duplicate records were identified and handled. The Date column was converted to datetime format and the Amount column was cleaned and converted to numeric format. After preprocessing, the dataset became clean, consistent, and suitable for further analysis.
